```python
import pandas as pd
import numpy as np
```

```python
path = r'C:\Users\All In One\Desktop\Automatização top sku\Shopping - Top Sku.csv'
df= pd.read_csv(path, skiprows=2)
df.head()
```

```python
df.replace(regex=r',',value='', inplace=True)
df.replace(regex=r'--',value=0, inplace=True)
df.head()
```

```python
df.columns
for column in df.columns:
    teste = pd.to_numeric(df[column], errors="ignore")
    df[column] = teste
df.dtypes
```

```python
df['ROI'] = 100 * df['Cost'].astype('float')/df['All conv. value'].astype('float')
pd.set_option('mode.use_inf_as_na', True)
df['ROI']=df['ROI'].replace(np.nan,0)
df.head()
```

```python
dfd = df.drop(columns=['Campaign','Currency code','CTR','Avg. CPC','Conv. rate'])
dfr = dfd.rename(columns={r'Item ID':'SKU','Clicks':'Cliques','Impr.':'Impressões','Cost':'Custo',\
                         'All conv. value':'Valor de pedidos', 'Conversions':'Conversões','Cost / conv.':'Custo/Pedido'})
dfr.sort_values(['Conversões'],ascending=False).round(2).head(20)
```

```python
#get period name
path = r'C:\Users\All In One\Desktop\Automatização top sku\Shopping - Top Sku.csv'
df= pd.read_csv(path, nrows=1)
df['Shopping - Top Sku'].replace(['January', 'February', 'March', 'April', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], \
                                 ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],\
                                 regex=True, inplace=True)
nome = str(df[['Shopping - Top Sku']])
nome = nome[37:150]
```

```python
out_path = r'C:\Users\All In One\Desktop\Automatização top sku\Top Sku Resultado ' + nome + '.xlsx'
dfr.sort_values(['Conversões'],ascending=False).round(2).to_excel(out_path,encoding='utf-8-sig',index=False)
print(out_path)
```
